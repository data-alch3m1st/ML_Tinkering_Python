```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import spacy

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC

from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    classification_report, accuracy_score
    , confusion_matrix, ConfusionMatrixDisplay
)

import warnings
warnings.simplefilter('ignore')

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
# Load SpaCy model

nlp = spacy.load("en_core_web_sm")
```

```python
# UPDATED Function ~ spaCy model (use 'en_core_web_sm' for speed, or 'lg' if you must)

nlp = spacy.load('en_core_web_sm', disable=['ner', 'parser'])  # Only keep tagger, lemmatizer

# Preprocess tweets: lemmatize, remove stopwords and punct, join as string
def preprocess(texts):
    docs = nlp.pipe(texts, batch_size=128, n_process=1)  # Use batch processing for speed
    return [
        " ".join([token.lemma_ for token in doc if not token.is_stop and not token.is_punct])
        for doc in docs
    ]
```

```python
# Load the dataset from Hugging Face:
df = pd.read_parquet("hf://datasets/tdavidson/hate_speech_offensive/data/train-00000-of-00001.parquet")

df.shape
df.head()
df.info()
```

```python
df['class'].value_counts()
```

```python
# Encode labels:
# 0: 'hate-speech', 1: 'offensive-language' or 2: 'neither'

df['label_actual'] = df['class'].map({0: 'hate-speech', 1: 'offensive-language', 2: 'neither' })
```

```python
# Visualize label distribution
sns.countplot(
    x='class', data=df, hue='label_actual'
)
plt.title("Label Distribution")
plt.show();
```

```python
df.head(1)
```

```python
X = df['tweet']
y = df['class']

X.shape, y.shape
```

```python
# Preprocess all tweets ONCE

X_processed = preprocess(X)
```

```python
# Split dataset (Now that we've absorbed the pre-processed X using our function, there will be much less work on the backent!)
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, test_size=0.2, random_state=27, stratify=y
)

# X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# # Just setting up pipes for future mod (gotta check my notes to make sure I do it right!!!)

model_pipes = Pipeline([
    ('tfidf', TfidfVectorizer(lowercase=True))
    , ('base_model', LogisticRegression(random_state=27)),
])
```

```python
# Grid CV for hyperparams fine-tuning;
param_grid = {
    'base_model__C': np.logspace(-2,2,13)  # Regularization 0.01-100 (numpy for perf log10's + 2 in-betweens;)
    , 'base_model__penalty': ['l2']                       # L1/L2 penalty (L2 most common with tfidf/BOW)
    , 'base_model__solver': ['lbfgs', 'saga']             # Solvers that support multiclass
    , 'base_model__multi_class': ['multinomial', 'ovr']   # Explicit multi-class strategies
    , 'base_model__max_iter': [1000]
}

grid_search = GridSearchCV(
    estimator=model_pipes
    , param_grid=param_grid
    , cv=3                  # Reduced from 5 to 3 folds given this dataset is a beast...
    , scoring='accuracy'
    , n_jobs=-1
)
```

```python
# Fit that ish!

grid_search.fit(X_train, y_train)
```

```python
# Print best hyperparameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)
```

```python
# Evaluate best model on test data
best_model = grid_search.best_estimator_

# Now FIT the best_model (MUST DO BEFORE PREDICTING!!! {otherwise, you essentially reset the model! :o })
best_model.fit(X_train, y_train)
```

```python
y_test_pred = best_model.predict(X_test)

print("Test Accuracy:", accuracy_score(y_test, y_test_pred))
confusion_matrix(y_test, y_test_pred)
print(classification_report(y_test, y_test_pred))
```

```python
ConfusionMatrixDisplay.from_estimator(
    best_model, X_test, y_test, normalize='true'
    , cmap='Blues', values_format='.4f'
);
```
