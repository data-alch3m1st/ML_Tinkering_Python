```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.pipeline import Pipeline

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# _jupyt3rNotebooks/py7Edu_learningMaterials/inspired_tinkering/data

df = pd.read_csv('../data/gene_expression.csv')
df.head()
```

```python
# Features and target
X = df.drop('Cancer Present', axis=1)
y = df['Cancer Present']

X.shape, y.shape
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20
    , random_state=27
)
```

```python
# Define pipeline steps
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier())
])
```

```python
# Parameter grid for n_neighbors
param_grid = {'knn__n_neighbors': 
              np.arange(1, 21)}

# GridSearchCV
grid = GridSearchCV(
    estimator=pipe
    , param_grid=param_grid
    , cv=5
    , scoring='accuracy'
    , return_train_score=True
)

grid.fit(X_train, y_train)
```

```python
neighbors = param_grid['knn__n_neighbors']
test_scores = grid.cv_results_['mean_test_score']

plt.figure(figsize=(8,5))
plt.plot(
    neighbors, 1 - test_scores
    , marker='o'
    , label='Error Rate (1 - accuracy)')

plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Score')
plt.title('KNN Error Rate vs. k')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    neighbors
    , test_scores
    , marker='s'
    , label='Accuracy'
)
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Score')
plt.title('KNN Accuracy vs. k')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Best n_neighbors
best_k = grid.best_params_['knn__n_neighbors']
print(f'Best n_neighbors: {best_k}')
print(f'Best cross-validated accuracy: {grid.best_score_:.3f}')

# Use best estimator for predictions
best_model = grid.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
print("Test Set Accuracy:"
      , accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
print(classification_report(y_test, y_pred))
```

```python
sns.heatmap(
    cm, annot=True, fmt="d"
)
```
