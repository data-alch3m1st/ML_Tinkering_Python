```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import (
    confusion_matrix
    , classification_report
    , ConfusionMatrixDisplay
    , roc_auc_score
    , RocCurveDisplay
    , PrecisionRecallDisplay
)

import warnings
warnings.simplefilter('ignore')

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
# Load the Titanic dataset
df = sns.load_dataset('titanic')

# Display the first few rows of the dataset
df.head()
df.info()
```

```python
# Drop rows with missing values in critical columns
df = df.dropna(
    subset=['age', 'embarked', 'fare'
            , 'sex', 'pclass'])

# Encode categorical variables into numerical format
df['sex'] = df['sex'].map(
    {'male': 0, 'female': 1}
)

df['embarked'] = df['embarked'].map(
    {'C': 0
     , 'Q': 1
     , 'S': 2}
)
```

```python
# Select features (X) and target variable (y)
X = df[['pclass', 'sex', 'age', 'fare', 'embarked']]
y = df['survived']

# Display the shape of the features and target
X.shape, y.shape
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=27
)

# Display the shapes of the training and test sets
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform both training and test data
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
# Initialize the SVC model with default parameters
svc = SVC()

# Fit the SVC model on the training data
svc.fit(scaled_X_train, y_train)

# Evaluate the model on the test data
base_accuracy = svc.score(scaled_X_test, y_test)

# Display the base accuracy (optional for understanding but avoid excessive printing)
base_accuracy
```

```python
# Eval performance of base (for comparison later);

y_pred_base = svc.predict(scaled_X_test)
score_base = accuracy_score(y_test, y_pred_base)
score_base
```

```python
# This logspace formula will give all the 'perfect' log10's from 0.001-100 w/ 2 nums in between each;
# np.logspace(-3,2,16).tolist()
```

```python
# Define the parameter grid for GridSearchCV
param_grid = {
    'C': np.logspace(-3, 2, 16)
    , 'kernel': ['rbf'] # ['linear', 'rbf', 'poly'] - all are available but rbf is usually way to go;
    , 'gamma': ['scale', 'auto']
  #  , 'degree': [2, 3, 4]  # Only relevant for 'poly' kernel
}

# Initialize the GridSearchCV object
grid_search = GridSearchCV(
    estimator=svc
    , param_grid=param_grid
    , scoring='accuracy'
    , cv=5
)
```

```python
# Fit GridSearchCV on the training data * Note: I usually run 'fit' in a sep cell from GridSearchCV / params in case of issues w/ the fit or I have to tweak params;
grid_search.fit(scaled_X_train, y_train)
```

```python
# Get the best hyperparameters and the best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Display the best parameters (optional)
best_params, best_score
```

```python
# Initialize and fit the 'best model' then see how it does on test data:


best_model = grid_search.best_estimator_
best_model.fit(scaled_X_train, y_train)
best_model.score(scaled_X_test, y_test)
```

```python
from sklearn.metrics import (
    confusion_matrix
    , classification_report
    , ConfusionMatrixDisplay
    , roc_auc_score
    , RocCurveDisplay
    , PrecisionRecallDisplay
)
```

```python
# Predict on test set
y_pred = best_model.predict(scaled_X_test)
score_best = accuracy_score(y_test, y_pred)
score_best
```

```python
# Run a quick comparison on best vs base (scores based on predictions;)

print(f"Best Model vs. Base Model : { np.round((((score_best - score_base) / score_base) * 100), 4)}%" )
```

```python
# Comparison of CMs (text/array, non-viz) for Best vs. Base:

print("Confusion Matrix, Best Model:")
confusion_matrix(y_test, y_pred)
print('\n')
print("Confusion Matrix, Base Model:")
confusion_matrix(y_test, y_pred_base)
```

```python
# Comparison of classification reports for Best vs. Base:

print("Classification Report, Best Model:")
print(classification_report(y_test, y_pred))
print('\n')
print("Classification Report, Base Model:")
print(classification_report(y_test, y_pred_base))
```

```python
# To do a side-by-side, normalized CM comparison (Best vs. Base):

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ConfusionMatrixDisplay.from_estimator(
    best_model, scaled_X_test, y_test, cmap='Blues', ax=ax[0]
)

ConfusionMatrixDisplay.from_estimator(
    svc, scaled_X_test, y_test, cmap='Reds', ax=ax[1]
)

ax[0].set_title('Best Model')
ax[1].set_title('Base Model')

plt.tight_layout()
plt.show();
```

```python
# To do a side-by-side, normalized CM comparison (Best vs. Base):

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ConfusionMatrixDisplay.from_estimator(
    best_model, scaled_X_test, y_test
    , normalize='true', cmap='Blues'
    , values_format='.4f'
    , ax=ax[0]
)

ConfusionMatrixDisplay.from_estimator(
    svc, scaled_X_test, y_test
    , normalize='true', cmap='Reds'
    , values_format='.4f'
    , ax=ax[1]
)

ax[0].set_title('Best Model')
ax[1].set_title('Base Model')

plt.tight_layout()
plt.show();
```

```python
# To do a side-by-side, normalized PRC comparison (Best vs. Base):

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

PrecisionRecallDisplay.from_estimator(
    best_model, scaled_X_test, y_test, ax=ax[0]
)

PrecisionRecallDisplay.from_estimator(
    svc, scaled_X_test, y_test, ax=ax[1]
)

ax[0].set_title('Best Model (Precision Recall Curve)')
ax[1].set_title('Base Model (Precision Recall Curve)')

plt.tight_layout()
plt.show();
```

```python
# To do a side-by-side, normalized ROC/AUC comparison (Best vs. Base):

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

RocCurveDisplay.from_estimator(
    best_model, scaled_X_test, y_test, ax=ax[0]
)

RocCurveDisplay.from_estimator(
    svc, scaled_X_test, y_test, ax=ax[1]
)

ax[0].set_title('Best Model (ROC/AUC Curve)')
ax[1].set_title('Base Model (ROC/AUC Curve)')

plt.tight_layout()
plt.show();
```

```python
# Another option - condense the 2x PRC and 2x ROC/AUC curve(s) plots into
# 2x subplots, w/ best/base as sep curves on a single plot;

fig, ax = plt.subplots(1, 2, figsize=(12, 8))

# Plot PRC curve for the best model
PrecisionRecallDisplay.from_estimator(
    best_model, scaled_X_test, y_test, ax=ax[0], name="Best Model"
)

# Plot PRC curve for the base model
PrecisionRecallDisplay.from_estimator(
    svc, scaled_X_test, y_test, ax=ax[0], name="Base Model"
)

# Plot ROC curve for the best model
RocCurveDisplay.from_estimator(
    best_model, scaled_X_test, y_test, ax=ax[1], name="Best Model"
)

# Plot ROC curve for the base model
RocCurveDisplay.from_estimator(
    svc, scaled_X_test, y_test, ax=ax[1], name="Base Model"
)

# Add title(s) and legend(s):
ax[0].set_title('PRC Comparison: Best Model vs Base Model')
ax[1].set_title('ROC/AUC Comparison: Best Model vs Base Model')
ax[0].legend(loc="lower left")
ax[1].legend(loc="lower right")

plt.tight_layout()
plt.show();
```
